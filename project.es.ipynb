#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_cinturon_sur_Madrid=["Fuenlabrada","Leganés","Getafe","Alcorcón"]
subconjunto_ds=ds[ds["level5"].isin(poblaciones_cinturon_sur_Madrid)]

subconjunto_ds.loc[:,"pps"]=subconjunto_ds["price"] / subconjunto_ds ["surface"]

pps_Getafe=subconjunto_ds[subconjunto_ds["level5"]=="Getafe"]["pps"].mean()
pps_Alcorcón=subconjunto_ds[subconjunto_ds["level5"]=="Alcorcón"]["pps"].mean()

print("Getafe:",pps_Getafe)

print("Alcorcón:",pps_Alcorcón)
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_cinturon_sur_Madrid=["Fuenlabrada","Leganés","Getafe","Alcorcón"]
subconjunto_ds=ds[ds["level5"].isin(poblaciones_cinturon_sur_Madrid)]

import matplotlib.pyplot as plt

fig, axs =plt.subplots(2,2,figsize=(9,7))
axs=axs.flatten()

for i, poblacion in enumerate(poblaciones_cinturon_sur_Madrid):
    datos_para_graficar=subconjunto_ds[subconjunto_ds["level5"]==población]
    axs[i].scatter(datos_para_graficar["surface"],datos_para_graficar["price"],alpha=0.3)
    axs[i].set_title(poblacion)
    axs[i].set_xlabel("Superficie (m2)")
    axs[i].set_ylabel("Precio")

plt.suptitle("Comparación de precios por m2 de cinturón sur de Madrid")
plt.tight_layout()
plt.show()
```
